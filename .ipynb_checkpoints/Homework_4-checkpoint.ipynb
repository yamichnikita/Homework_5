### Информация о признаках:

1) Идентификационный номер


2) Диагноз (M = злокачественный, B = доброкачественный)
3-32)


Для каждого ядра клетки вычисляются десять характеристик:


а) радиус (среднее значение расстояний от центра до точек по периметру)


б) текстура (стандартное отклонение значений шкалы серого)


c) периметр


d) площадь


e) плавность (локальное изменение длины радиуса)


f) компактность (периметр^2 / площадь - 1,0)


g) вогнутость (выраженность вогнутых участков контура)


h) вогнутые точки (количество вогнутых участков контура)


i) симметрия


j) фрактальная размерность ("приближение береговой линии" - 1)

Среднее значение, стандартная ошибка и "наихудшее" или наибольшее (среднее из трех наибольших значений) этих признаков были вычислены для каждого изображения, в результате чего было получено 30 признаков. Например, поле 3 - Средний радиус, поле 13 - Радиус SE, поле 23 - Наихудший радиус.

### Посторим гистограммы/распределения

```python
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
sns.histplot(data= df, x="diagnosis", )
```

```python
Выясним, как диагноз зависит от среднего значения радиуса клетки.
```

```python
diagnosis_to_radius_mean = df.groupby('diagnosis')['radius_mean'].mean()
diagnosis_to_radius_mean
```

```python
diagnosis_to_radius_mean.plot.bar();
```

На гистограмме видно, что у злокачественной клетки радиус, в среднем, больше чем у доброкачественной.

### Построим heatmap для матрицы корреляций

```python
plt.subplots(figsize = (18, 15))
sns.heatmap(df.corr(), cmap = sns.color_palette('coolwarm', 10000), vmin = -1, center =0)
plt.show()
```

Так как параметры mean, se, worst скоррелированы между собой, то выберем что-то одно из них и постороим матрицу корреляций ещё раз. Например, параметры средних значений - mean.

```python
df1 = df.iloc[:, 2:12]
```

```python
df1
```

```python
plt.subplots(figsize = (10, 7))
sns.heatmap(df1.corr(), cmap = sns.color_palette('coolwarm', 10000), vmin = -1, center =0, annot = True)
plt.show()
```

Из матрицы корреляций видно, что между собой сильно скоррелированы радиус (radius_mean), периметр (perimetr_mean) и площадь клетки (area_mean). Из-за того, что площадь и периметр математически выражаются через радиус. Менее скоррелирован параметр вогнутых точек (concave points_mean) и радиуса (radius_mean), параметр вогнутых точек (concave points_mean) и вогнутость (concavity_mean), параметр вогнутых точек (concave points_mean) и компакнтность (compactness_mean).

Возможно, не стоит использовать при обучении модели парметры периметра и площади клетки, т.к. они в прямой зависимости от радиуса клетки.
